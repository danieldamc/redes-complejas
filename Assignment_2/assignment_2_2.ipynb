```python
import networkx as nx
import numpy as np

small_graph = nx.read_pajek('data/redchica.net')

degree_dict = dict(small_graph.degree())
betweenness_dict = nx.betweenness_centrality(small_graph)
pagerank_dict = nx.pagerank(small_graph)

# print("Degree of each node:")
# for node, degree in degree_dict.items():
#     print(f"Node {node}: {degree}")

# print("\nBetweenness centrality of each node:")
# for node, betweenness in betweenness_dict.items():
#     print(f"Node {node}: {betweenness}")

# print("\nPageRank of each node:")
# for node, pagerank in pagerank_dict.items():
#     print(f"Node {node}: {pagerank}")

labels = {node: f"Degree: {degree_dict[node]}\nBetweenness: {betweenness_dict[node]:.2f}\nPageRank: {pagerank_dict[node]:.2f}"
          for node in small_graph.nodes()}

#pos = nx.spring_layout(small_graph)
pos = {
    'v1': np.array([ 0.34962515, -0.88787688]), 
    'v2': np.array([-0.41715934, -0.68139464]), 
    'v3': np.array([ 0.03307588, -0.16499241]), 
    'v4': np.array([-0.3129866 ,  0.23604149]), 
    'v5': np.array([-0.00562093,  0.59102928]), 
    'v6': np.array([0.43195524, 0.14514059]), 
    'v7': np.array([-0.8663992,  .3       ]), 
    'v8': np.array([ 0.78750979, -0.23794743])}
print(pos)
#set plot size
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
nx.draw(small_graph, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray')

#nx.draw_networkx_nodes(small_graph, pos, node_color='lightblue', node_size=3000)
#nx.draw_networkx_edges(small_graph, pos, edge_color='gray', width=2)
label_pos = {node: (x, y + 0.1) for node, (x, y) in pos.items()}  # Offset on the y-axis
nx.draw_networkx_labels(small_graph, label_pos, labels=labels, horizontalalignment='center', font_size=5, font_color='black');
plt.show()
```
